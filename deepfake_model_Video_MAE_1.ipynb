```python
from transformers import pipeline, AutoImageProcessor, AutoModelForVideoClassification
import torch
import decord

# Pipeline rápido
pipe = pipeline("video-classification", model="Hemgg/deepfake_model_Video-MAE-1", device=0)

# Modelo manual
processor = AutoImageProcessor.from_pretrained("Hemgg/deepfake_model_Video-MAE-1")
model = AutoModelForVideoClassification.from_pretrained("Hemgg/deepfake_model_Video-MAE-1").to("cuda")
```

```python
from google.colab import files

# Escolher arquivo do computador
uploaded = files.upload()

# Pega o nome do primeiro arquivo enviado
video_path = list(uploaded.keys())[0]
print("Vídeo carregado:", video_path)
```

```python
results = pipe(video_path)
print(results)
```

```python
best = max(results, key=lambda x: x['score'])

if best['label'].lower() == "fake" and best['score'] > 0.7:
    print("⚠️ Provável deepfake!")
else:
    print("✅ Provavelmente real")
```
